### 1️⃣ Formulate at least two hypotheses from the dataset. Example:
* H₀: Smoking has no effect on Diabetes prevalence.
* H₁: Smoking affects Diabetes prevalence.


```python
table1 = pd.crosstab(df['smoking_status'], df['diabetes'])
table1
```

```python
chi2, p_value, dof, expected = chi2_contingency(table1)
print("Chi-square value:", chi2)
print("P-value:", p_value)

print("---------------------------------------------------------------")

if p_value < 0.05:
    print("Reject H0: Smoking status affects diabetes prevalence.")
else:
    print("Accept H0: Smoking status has no significant effect on diabetes.")
```

#### 📝 Interpretation:
The p-value (**0.701**) is much greater than 0.05, and the chi-square value (**0.71**) is far below the critical value (**5.99**). This means there is **no significant relationship** between smoking status and diabetes in this sample.

#### ✅ Conclusion:
We **accept H₀** — smoking status has no significant effect on diabetes prevalence. Diabetes occurs at similar rates regardless of smoking habit.


#### 🧩 Perform ANOVA Test

```python
daily = df[df['exercise_frequency']=="Daily"]['bmi']
weekly = df[df['exercise_frequency']=="Weekly"]['bmi']
rarely = df[df['exercise_frequency']=="Rarely"]['bmi']
never = df[df['exercise_frequency']=="Never"]['bmi']
```

```python
groups = [daily, weekly, rarely, never]

k = len(groups)        # number of groups

N = sum(len(group) for group in groups)

print("Number of groups:", k)
print("Total observations:", N)
```

```python
all_bmi = pd.concat(groups)

overall_mean = all_bmi.mean()

print("Overall Mean BMI:", overall_mean)
```

```python
SSB = 0

for group in groups:
    group_size = len(group)
    group_mean = group.mean()
    
    SSB += group_size * (group_mean - overall_mean)**2

print("Between Group Sum of Squares:", SSB)
```

```python
SSW = 0

for group in groups:
    group_mean = group.mean()
    
    SSW += sum((group - group_mean)**2)

print("Within Group Sum of Squares:", SSW)
```

```python
df_between = k - 1

df_within = N - k

print("DF Between:", df_between)
print("DF Within:", df_within)
```

```python
MSB = SSB / df_between

MSW = SSW / df_within


print("Mean Square Between:", MSB)

print("Mean Square Within:", MSW)
```

```python
F_value = MSB / MSW

print("F Statistic:", F_value)
```

```python
p_value = 1 - f.cdf(F_value, df_between, df_within)

print("P-value:", p_value)

alpha = 0.05

print("------------------------------------------------------------")

if p_value < alpha:
    print("Reject H0")
    print("Exercise frequency has a significant effect on BMI.")
    
else:
    print("Accept H0")
    print("Exercise frequency does not have a significant effect on BMI.")
```

#### 📝 Interpretation:
The F-value (**0.87**) is below the critical value (**2.62**), and the p-value (**0.458**) is well above 0.05. The differences in mean BMI across exercise groups (Daily, Weekly, Rarely, Never) are **not statistically significant**.

#### ✅ Conclusion:
We **accept H₀** — exercise frequency does not have a significant effect on BMI in this dataset.


### 2️⃣ Calculate Confidence Intervals for key numerical data like age, weight, etc.

```python
columns = [
    'age',
    'weight',
    'blood_pressure',
]
```

```python
confidence = 0.95

for column in columns:
    
    data = df[column]
    
    # Number of observations
    n = len(data)
    
    # Mean
    mean = data.mean()
    
    # Standard deviation
    std = data.std()
    
    # Standard error
    standard_error = std / np.sqrt(n)
    
    # t critical value
    t_value = t.ppf((1 + confidence) / 2, n-1)
    
    # Margin of error
    margin_error = t_value * standard_error
    
    # Confidence interval
    lower_limit = mean - margin_error
    upper_limit = mean + margin_error
    
    print("\nColumn:", column)
    print("Mean:", round(mean,2))
    print("95% Confidence Interval:",
          round(lower_limit,2),
          "to",
          round(upper_limit,2))
```

#### 📝 Interpretation:
We are **95% confident** the true population mean lies within these ranges: Age between **47.24 and 50.48**, Weight between **79.42 and 83.37**, Blood Pressure between **135.42 and 139.56**.

#### ✅ Conclusion:
Since all intervals are fairly narrow, the estimates are **precise and reliable**. The average age is approximately **49 years**, weight approximately **81 kg**, and blood pressure approximately **137**, with only minor variation.


###  3️⃣ Find the Critical Value and p-value to interpret the test results.


```python
from scipy.stats import f

alpha = 0.05

df_between = 3
df_within = len(df)-4

critical_value = f.ppf(
    1-alpha,
    df_between,
    df_within
)

p_value = 1 - f.cdf(
    F_value,
    df_between,
    df_within
)

print("Critical Value:", critical_value)
print("P-value:", p_value)

print("-------------")

if p_value < alpha:
    print("Reject H0")
else:
    print("Accept H0")
```

#### 📝 Interpretation:
Using the critical value approach confirms the same result: F (**0.87**) < critical value (**2.62**), and p (**0.458**) > 0.05.

#### ✅ Conclusion:
We **accept H₀** — this re-confirms exercise frequency has no significant effect on BMI.


### 4️⃣ Perform z-test or t-test based on sample size (mean comparison across groups).


```python
male_bmi = df[df['gender']=="Male"]['bmi']

female_bmi = df[df['gender']=="Female"]['bmi']
```

```python
n1 = len(male_bmi)
n2 = len(female_bmi)

print("Male sample size:", n1)
print("Female sample size:", n2)
```

```python
mean1 = male_bmi.mean()

mean2 = female_bmi.mean()

print("Male BMI Mean:", mean1)

print("Female BMI Mean:", mean2)
```

```python
var1 = male_bmi.var()

var2 = female_bmi.var()
```

```python
standard_error = np.sqrt(
    (var1/n1) + (var2/n2)
)
```

```python
t_value = (mean1 - mean2) / standard_error

print("T Statistic:", t_value)
```

```python
df_value = n1 + n2 - 2

print("Degrees of Freedom:", df_value)
```

```python
p_value = 2 * (1 - t.cdf(abs(t_value), df_value))

print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
    print("There is a significant difference between male and female BMI.")
    
else:
    print("Accept H0")
    print("There is no significant difference between male and female BMI.")
```

#### 📝 Interpretation:
The p-value (**0.095**) is greater than 0.05, and the t-value (**1.67**) is below the critical value (**1.97**). The slightly higher male BMI (**29.37**) vs female BMI (**27.62**) is **not statistically significant**.

#### ✅ Conclusion:
We **accept H₀** — there is no significant difference in BMI between males and females.


### 5️⃣ Conduct a chi-square test on categorical data (e.g., Smoking habit vs Disease).

```python
table = pd.crosstab(
    df['smoking_status'],
    df['diabetes']
)

table
```

```python
row_total = table.sum(axis=1)

column_total = table.sum(axis=0)

total = table.sum().sum()


expected = np.outer(row_total, column_total) / total


expected = pd.DataFrame(
    expected,
    index=table.index,
    columns=table.columns
)


expected
```

```python
chi_square_value = ((table - expected)**2 / expected).sum().sum()

print("Chi-square value:", chi_square_value)
```

```python
rows = table.shape[0]

columns = table.shape[1]


df_value = (rows-1)*(columns-1)


print("Degrees of Freedom:", df_value)
```

```python
p_value = chi2.sf(
    chi_square_value,
    df_value
)
```

```python
alpha = 0.05
critical_value = chi2.ppf(
    1-alpha,
    df_value
)


print("Critical Value:", critical_value)
```

```python
if p_value < alpha:
    
    print("Reject H0")
    print("Smoking status and diabetes are significantly related.")

else:
    
    print("Accept H0")
    print("Smoking status and diabetes are independent.")
```

#### 📝 Interpretation:
Same as the earlier chi-square result: chi-square (**0.71**) is far below the critical value (**5.99**), p-value (**0.701**) is well above 0.05.

#### ✅ Conclusion:
We **accept H₀** — smoking status and diabetes are independent of each other.


###  6️⃣ Perform an ANOVA test to check if age groups significantly differ in disease rate.

```python
df['disease_rate'] = df['diabetes'].astype(int)
```

```python
age_groups = [
    df[df['age_group']=="18-25"]['disease_rate'],
    df[df['age_group']=="26-35"]['disease_rate'],
    df[df['age_group']=="36-45"]['disease_rate'],
    df[df['age_group']=="46-60"]['disease_rate'],
    df[df['age_group']=="60+"]['disease_rate']
]
```

```python
k = len(age_groups)

N = sum(len(group) for group in age_groups)


print("Number of groups:", k)

print("Total observations:", N)
```

```python
all_data = pd.concat(age_groups)

overall_mean = all_data.mean()


print("Overall Disease Rate:", overall_mean)
```

```python
SSB = 0

for group in age_groups:
    
    n = len(group)
    
    group_mean = group.mean()
    
    SSB += n * (group_mean - overall_mean)**2


print("SSB:", SSB)
```

```python
SSW = 0


for group in age_groups:
    
    group_mean = group.mean()
    
    SSW += sum((group - group_mean)**2)


print("SSW:", SSW)
```

```python
df_between = k-1

df_within = N-k


print(df_between)

print(df_within)
```

```python
MSB = SSB / df_between

MSW = SSW / df_within


print("MSB:", MSB)

print("MSW:", MSW)
```

```python
F_value = MSB / MSW


print("F-value:", F_value)
```

```python
alpha = 0.05


critical_value = f.ppf(
    1-alpha,
    df_between,
    df_within
)


print("Critical Value:", critical_value)
```

```python
p_value = 1 - f.cdf(
    F_value,
    df_between,
    df_within
)


print("P-value:", p_value)
```

```python
if p_value < alpha:
    
    print("Reject H0")
    print("Disease rate differs significantly among age groups.")

else:
    
    print("Accept H0")
    print("Disease rate does not differ significantly among age groups.")
```

#### 📝 Interpretation:
The F-value (**1.50**) is below the critical value (**2.39**), and the p-value (**0.201**) is greater than 0.05. Although diabetes rate ranges from **66.7% to 80.8%** across age groups, this spread is **not statistically significant**.

#### ✅ Conclusion:
We **accept H₀** — diabetes rate does not differ significantly across age groups.


### 7️⃣ Calculate Covariance and Correlation between continuous variables (e.g., Age vs. BMI)

```python
age = df['age']
bmi = df['bmi']
```

```python
age_mean = age.mean()
bmi_mean = bmi.mean()

print("Mean Age:", age_mean)
print("Mean BMI:", bmi_mean)
```

```python
n = len(df)


covariance = sum(
    (age - age_mean) *
    (bmi - bmi_mean)) / (n-1)


print("Covariance:", covariance)
```

```python
age_std = age.std()

bmi_std = bmi.std()
```

```python
correlation = covariance / (age_std * bmi_std)


print("Correlation:", correlation)
```

#### 📝 Interpretation:
The correlation coefficient (**r = 0.109**) indicates a **very weak positive relationship** between age and BMI — age explains only about 1% of the variation in BMI.

#### ✅ Conclusion:
Age and BMI are **essentially unrelated** in this dataset; growing older does not meaningfully predict higher BMI here.
